```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
company_names = ['Maruti Suzuki', 'Hyundai', 'Kia', 'Mahindra', 'Tata', 'Toyota', 'Renault', 'MG', 'Ford', 'Nissan', 'Honda']

market_share_2020 = [24.87, 18.69, 8.98, 18.97, 6.28, 7.07, 4.52, 2.32, 4.69, 0.55, 1.56]
market_share_2021 = [21.60, 20.19, 14.68, 14.66, 8.15, 6.25, 4.85, 3.36, 3.31, 1.08, 0.91]
```

```python
x = np.arange(0, len(company_names))
width = 0.3
```

```python
fig, ax = plt.subplots()
rects1 = ax.bar(x - width / 2, market_share_2020, width, label = 'Market Share 2020')
rects2 = ax.bar(x + width / 2, market_share_2021, width, label = 'Market Share 2021')

ax.set_ylabel('Percent Share')
ax.set_title('Market Shares in year 2020 and 2021')
ax.set_xticks(x)
ax.set_xticklabels(company_names, rotation = 45)
ax.legend()
fig.tight_layout()
plt.savefig('finlatics_sector_1_rivalry.png', dpi = 300)

plt.show()
```
