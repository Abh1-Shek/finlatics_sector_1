```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
company_names = ['Hero Motocorp', 'Honda', 'TVS', 'Bajaj', 'Royal-Enfield']
```

```python
market_share_2019 = [38.74, 24.47, 14.07, 11.97, 3.77]
market_share_2020 = [39.25, 24.55, 14.35, 11.30, 3.58]
```

```python
x = np.arange(0, len(company_names))
width = 0.3
```

```python
fig, ax = plt.subplots()
rects1 = ax.bar(x - width / 2, market_share_2019, width, label = 'Market Share 2019')
rects2 = ax.bar(x + width / 2, market_share_2020, width, label = 'Market Share 2020')

ax.set_ylabel('Percent Share')
ax.set_title('Market Shares in year 2019 and 2020')
ax.set_xticks(x)
ax.set_xticklabels(company_names, rotation = 45)
ax.legend()
fig.tight_layout()
plt.savefig('finlatics_sector_1_rivalry.png', dpi = 300)

plt.show()
```
