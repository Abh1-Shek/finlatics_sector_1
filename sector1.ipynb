```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
company_name = ['Maruti', 'Hyundai', 'Tata', 'Kia', 'Mahindra', 'Renault', 'Toyota', 'Honda']
market_share = [50, 17.4, 7, 5.8, 5.6, 3.3, 3.1, 2.9]
```

```python
plt.pie(market_share, labels = company_name, autopct = '%.1f%%', radius = 1)
plt.savefig('finlatics_automotive_market_share.png', dpi = 500)
plt.show()
```
